```python
import pandas as pd
import numpy as np
import sys
sys.path.append('..')
from src.transformation.feature_eng import FeatureEngineer
from src.transformation.transformer import DataTransformer
print('✅ Pronto!')
```

## 📥 1. Dataset

```python
np.random.seed(42)
df = pd.DataFrame({
    'data':    pd.date_range('2024-01-01', periods=200, freq='D'),
    'regiao':  np.random.choice(['SP','RJ','MG','RS'], 200),
    'produto': np.random.choice(['A','B','C'], 200),
    'valor':   np.random.uniform(100, 5000, 200).round(2),
    'qtd':     np.random.randint(1, 50, 200),
    'target':  np.random.randint(0, 2, 200)
})
df.head()
```

## 🗓️ 2. Features de data + coluna calculada

```python
df_t = (
    DataTransformer(df)
    .add_date_parts('data')
    .add_calculated_column('receita', 'valor * qtd')
    .get_result()
)
df_t.head()
```

## 🏷️ 3. Encoding + Scaling + Split

```python
cols = ['regiao','produto','valor','qtd','receita','data_mes','data_trimestre','target']
fe = FeatureEngineer(df_t[cols])
fe.encode_categoricals(['regiao','produto'], method='label')
fe.scale(columns=['valor','qtd','receita'], method='standard')
X_train, X_test, y_train, y_test = fe.split(target_col='target')
print(f'Train: {X_train.shape} | Test: {X_test.shape}')
```

```python
import os
os.makedirs('../data/datasets', exist_ok=True)
X_train.to_csv('../data/datasets/X_train.csv', index=False)
X_test.to_csv('../data/datasets/X_test.csv', index=False)
y_train.to_csv('../data/datasets/y_train.csv', index=False)
y_test.to_csv('../data/datasets/y_test.csv', index=False)
print('✅ Datasets salvos!')
```
