## 📥 1. Dados brutos com problemas

```python
df_raw = pd.DataFrame({
    'data':   ['2024-01-01','2024-01-02',None,'2024-01-01','2024-01-03'],
    'valor':  [100, None, 300, 100, -999],
    'regiao': ['SP','RJ','MG','SP','XX'],
    'qtd':    [10, 20, None, 10, 5]
})
print('Nulos:', df_raw.isnull().sum().to_dict())
df_raw
```

## 🧹 2. Pipeline de limpeza

```python
df_clean = (
    DataCleaner(df_raw)
    .remove_duplicates()
    .fill_nulls(strategy='median')
    .normalize_dates(columns=['data'])
    .validate_schema(['data','valor','regiao','qtd'])
    .get_result()
)
df_clean
```

## ✅ 3. Antes x Depois

```python
print(f'ANTES → Linhas: {len(df_raw)} | Nulos: {df_raw.isnull().sum().sum()}')
print(f'DEPOIS → Linhas: {len(df_clean)} | Nulos: {df_clean.isnull().sum().sum()}')
```

```python
import os
os.makedirs('../data/processed', exist_ok=True)
df_clean.to_csv('../data/processed/dados_limpos.csv', index=False)
print('✅ Salvo em data/processed/dados_limpos.csv')
```
