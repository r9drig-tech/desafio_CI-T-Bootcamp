## 📊 Dataset de demonstração

```python
df = pd.DataFrame({
    'mes':      pd.period_range('2024-01', periods=12, freq='M').astype(str),
    'regiao':   ['SP','RJ','MG','RS','BA','PR','SC','GO','PE','CE','AM','PA'],
    'vendas':   [520,310,280,190,140,220,175,130,160,145,90,85],
    'meta':     [500,300,250,180,150,200,170,120,155,140,100,90],
    'clientes': [1200,800,700,500,380,580,460,340,420,380,240,220]
})
df['atingiu_meta'] = df['vendas'] >= df['meta']
df
```

## 💡 1. Insight Agent — Resumo Executivo

```python
from src.agents.insight_agent import InsightAgent

agent = InsightAgent()
result = agent.run(
    df,
    context='Desempenho comercial regional 2024',
    output='executive_summary'
)
print(result)
```

## 📌 2. Insight Agent — Bullet Points

```python
bullets = agent.run(df, context='Desempenho comercial 2024', output='bullet_points')
print(bullets)
```

## 📦 3. Insight Agent — JSON estruturado

```python
import json
json_result = agent.run(df, context='Desempenho comercial 2024', output='json')
data = json.loads(json_result)
print(json.dumps(data, ensure_ascii=False, indent=2))
```

## 🔧 5. LLM Integration — Funções utilitárias

```python
from src.ai.llm_integration import summarize, classify, extract_entities

texto = 'As vendas da região SP superaram a meta em 4% no mês de março, impulsionadas pelo lançamento do produto A.'

print('📝 Resumo:')
print(summarize(texto, max_words=30))

print('\n🏷️ Classificação:')
print(classify(texto, ['positivo','negativo','neutro','alerta']))

print('\n🔍 Entidades:')
print(json.dumps(extract_entities(texto), ensure_ascii=False, indent=2))
```
